# Lineal model Regression

For thi model i am going to use the diabetes data set which it will help us 
to explain a little bit about the distribution 

```python
#%pip install scikit-learn matplotlib

from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt
```

```python
diabetes = load_diabetes()

X = diabetes.data
y = diabetes.target

AGE = 0; SEX=1; MASS_INDEX=2; BLOOD_PRESSURE = 3

x = range(0,diabetes.data.shape[0])

#Let check how the data spread in a chart 
plt.figure(figsize=(30,8))
plt.plot(x, X[:,MASS_INDEX], 'y-', label='Mass index')
plt.plot(x, X[:,BLOOD_PRESSURE], 'b-', label='Blood pression')
plt.rcParams.update({'font.size':22})
plt.legend(prop={'size':25}); plt.grid(); plt.show()
```

```python
fig, axs = plt.subplots(1,2,figsize=(12,4))

for fig, feature, label in zip(range(2),[MASS_INDEX,AGE],['Mass index', 'Age']):
    axs[fig].scatter(X[:,feature], X[:,BLOOD_PRESSURE], c='g', marker="o")
    axs[fig].set_xlabel(label); axs[fig].set_ylabel('Blood Pressure')
    axs[fig].grid()

plt.show()
```
