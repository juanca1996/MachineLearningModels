# Make blobs 

Con este data set podemos probar tipos de distribucion gaussiana para modelar 
con respecto a las clusterizacion

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=200,n_features=2,centers=3,random_state=10)

X,y
```

```python
plt.figure(figsize=(4,4))

for target,color, marker in zip(range(3),['y','g','b'],['o','^','+']):
    plt.scatter(X[y==target,0], X[y==target,1],c=color,marker=marker,
                label='class '+format(target))
    
plt.xlabel('feature 1',fontsize=15)
plt.ylabel('feature 2',fontsize=15)
plt.title('3 clases scatter plot')
plt.legend(prop={'size':14}); plt.grid()
plt.grid(True)
plt.show()
```
