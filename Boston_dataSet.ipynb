# Boston Data set

Here is an example using boston data set, which we are going to manipulate
to create models or prediction given this data set

```python
#%pip install pandas
import pandas as pd

boston = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/refs/heads/master/BostonHousing.csv")

X = boston.drop("medv", axis=1)
y = boston["medv"]

```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import statistics
import numpy as np

x = range(0,X.shape[0],1)

CRIME = 0; ZN=1; INDUSTRI=2;  ROOM=5; DISTANCE=7

X = np.array(X)
y = np.array(y)

plt.figure(figsize=(30,8))

plt.plot(x,X[:,INDUSTRI],'y-',label='industri')
plt.plot(x,X[:,DISTANCE], 'b-',label='Distance')
plt.plot(x, X[:,ROOM],'c-',label='Rooms')
plt.rcParams.update({'font.size':18})
plt.legend(prop={'size':22}); plt.grid()

plt.show()

```

```python
mean = []; variance = []


for i in range(boston.drop("medv", axis=1).shape[1]):
    mean.append(statistics.mean(X[:,i]))
    variance.append(statistics.variance(X[:,i]))

for i in range(boston.drop("medv", axis=1).shape[1]):
    print(format(i) + " mean: {:.2f}".format(mean[i]) +
          ", variance: {:.2f}".format(variance[i]))
    
```

```python
fig,axs = plt.subplots(1,1,figsize=(7,4))


ax2 = fig.add_subplot(1, 2, 2, projection='3d')
ax2.scatter(X[:,DISTANCE], X[:,ROOM], y, c='b', marker='o')
ax2.set_xlabel('distance', fontsize=22)
ax2.set_ylabel('rooms',fontsize=22)
ax2.set_zlabel('Boston', fontsize=22)
```
